# Decision Tree Classification on Tennis Dataset

## Introduction
This notebook demonstrates the use of decision trees for classification using a tennis dataset. The dataset includes various weather conditions and a decision (Play/No Play) based on these conditions.

### Dataset Description
- **Outlook**: Sunny, Overcast, Rainy
- **Temperature**: Hot, Mild, Cool
- **Humidity**: High, Normal
- **Windy**: Yes, No
- **Play**: Decision to play tennis (Yes or No)

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.preprocessing import LabelEncoder

# Load the dataset
file_path = 'tennis.csv'
tennis_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
tennis_data.head()
```

## Data Preprocessing
Before applying a decision tree, we need to preprocess the data. This involves converting categorical variables into a format that can be used by the classifier.

```python
# Converting categorical variables to numeric
encoder = LabelEncoder()
for column in tennis_data.columns:
    tennis_data[column] = encoder.fit_transform(tennis_data[column])

# Display the transformed data
tennis_data.head()
```

## Building the Decision Tree Classifier
We will now build the decision tree classifier using the entire dataset.

```python
# Building the decision tree classifier
X = tennis_data.drop('Play', axis=1)
y = tennis_data['Play']
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X, y)
```

## Visualizing the Decision Tree
To better understand the decision-making process of the model, we will visualize the decision tree.

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Visualizing the decision tree
plt.figure(figsize=(12, 8))
plot_tree(decision_tree, feature_names=X.columns, class_names=['No', 'Yes'], filled=True)
plt.title('Decision Tree Visualization')
plt.show()
```

```python
# Decision Tree Visualization Using Tennis Dataset

# Introduction
# This notebook demonstrates the creation and visualization of a decision tree using a tennis dataset.

import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt

# Load the Dataset
file_path = 'tennis.csv'  # Replace with your dataset path
tennis_data = pd.read_csv(file_path)
print(tennis_data.head())

# Preprocess the Data
encoder = LabelEncoder()
for column in tennis_data.columns:
    tennis_data[column] = encoder.fit_transform(tennis_data[column])
print(tennis_data.head())

# Build the Decision Tree
X = tennis_data.drop('Play', axis=1)
y = tennis_data['Play']
decision_tree = DecisionTreeClassifier(random_state=42)
decision_tree.fit(X, y)

# Visualize the Decision Tree
plt.figure(figsize=(12, 8))
plot_tree(decision_tree, feature_names=X.columns, class_names=encoder.inverse_transform(decision_tree.classes_), filled=True, rounded=True)
plt.show()
```
